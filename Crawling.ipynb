```python
import numpy as np
import pandas as pd
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from time import sleep
import csv
import json
import datetime
from selenium.webdriver import ActionChains
from collections import deque
```

```python
f = open('apartment.csv', 'w', encoding='utf-8', newline='')
w = csv.writer(f)

option = webdriver.ChromeOptions()

url = 'https://m.land.naver.com/search'
sleep(2)
browser = webdriver.Chrome(options=option)
browser.get(url)
# //*[@id="_countContainer"]/a[1]
browser.find_element(By.ID, 'query').send_keys("local", Keys.ENTER)
# browser.find_element(By.CSS_SELECTOR, 'a').execute_script('arguments[0].click();', javascript)
# /html/body/div[1]/div/div[2]/div[2]/section/div[2]/div/a[1]
sleep(5)
browser.find_element(By.XPATH,'/html/body/div[1]/div/div[2]/div[2]/section/div[2]/div/a[1]').click()
WebDriverWait(option, 100)
# browser.execute_script('void(0)')
sleep(3)

```

```python
start = datetime.datetime.now()
end = start + datetime.timedelta(seconds=(60*10)) # 지정 시간(초) 동안 스크롤 내리기
itemlist= browser.find_element(By.CLASS_NAME,"article_box.article_box--sale")

while True:
    browser.execute_script("arguments[0].scrollBy(0,5000)", itemlist)
    sleep(1)
    if datetime.datetime.now() > end:
        break
```

```python
depth = browser.find_elements(By.CSS_SELECTOR, 'div.item_area._Listitem')
title = browser.find_elements(By.CSS_SELECTOR, 'div.title_area')
price = browser.find_elements(By.CSS_SELECTOR, 'div.price_area')
info = browser.find_elements(By.CSS_SELECTOR, 'div.information_area')
tag = browser.find_elements(By.CSS_SELECTOR, 'div.tag_area')
cp = browser.find_elements(By.CSS_SELECTOR, 'div.cp_area')
merit = browser.find_elements(By.CSS_SELECTOR, 'div.merit_area')
```

```python
list_title = []
list_price = []
list_info = []
list_tag = []
list_cp = []
list_merit = []
list_depth = []
```

```python
def append_crawling(data,listname):
    for i in data:
        listname.append(i.text)
```

```python
append_crawling(title,list_title)
append_crawling(price,list_price)
append_crawling(info,list_info)
append_crawling(tag,list_tag)
append_crawling(cp,list_cp)
append_crawling(merit,list_merit)
append_crawling(depth,list_depth)
```

```python
print(len(list_title))
print(len(list_price))
print(len(list_info))
print(len(list_tag))
print(len(list_cp))
print(len(list_merit))
print(len(list_depth))
```

```python
pd.DataFrame({
    'title' : list_title,
    'price' : list_price,
    'info' : list_info,
    'tag' : list_tag,
    'merit' : list_merit,
    'depth' : list_depth
})
```
